```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sqlite3


try:
    baglanti=sqlite3.connect('ulkeler.db')
except:
    print("Veritabanı bağlantısı yapılamadı.")
    

abd = baglanti.execute("Select Enflasyon from abd  ").fetchall()
england = baglanti.execute("Select Enflasyon from England  ").fetchall()
france = baglanti.execute("Select Enflasyon from Fransa  ").fetchall()
spain = baglanti.execute("Select Enflasyon from Spain  ").fetchall()
ukraine = baglanti.execute("Select Enflasyon from Ukrayna  ").fetchall()



abd1 = baglanti.execute("Select Issizlik from abd  ").fetchall()
england1 = baglanti.execute("Select Issizlik from England  ").fetchall()
france1 = baglanti.execute("Select Issizlik from Fransa  ").fetchall()
spain1 = baglanti.execute("Select Issizlik from Spain  ").fetchall()
ukraine1 = baglanti.execute("Select Issizlik from Ukrayna  ").fetchall()


 
class ulkeler:
    def abd():
        i = 0 
        ay = 1
        while i < 13:
            enflasyon = int(input(f"{ay}. Ayın enflasyon verisini giriniz: "))
            issizlik = int(input(f"{ay}. Ayın işsizlik verisini giriniz: "))
            try:
                baglanti.execute("INSERT INTO abd(Enflasyon,Issizlik) VALUES ('{}','{}')".format(enflasyon,issizlik))
                baglanti.commit()
            except:
                print("Veritabanı bağlantısında bir hata oluştu.")
            finally:
                print("Ekleme Başarılı")
            ay = ay + 1
            i = i + 1
    def england():
        i = 0 
        ay = 1
        while i < 13:
            enflasyon = int(input(f"{ay}. Ayın enflasyon verisini giriniz: "))
            issizlik = int(input(f"{ay}. Ayın işsizlik verisini giriniz: "))
            try:
                baglanti.execute("INSERT INTO England(Enflasyon,Issizlik) VALUES ('{}','{}')".format(enflasyon,issizlik))
                baglanti.commit()
            except:
                print("Veritabanı bağlantısında bir hata oluştu.")
            finally:
                print("Ekleme Başarılı")
            ay = ay + 1
            i = i + 1
    def france():
        i = 0 
        ay = 1
        while i < 13:
            enflasyon = int(input(f"{ay}. Ayın enflasyon verisini giriniz: "))
            issizlik = int(input(f"{ay}. Ayın işsizlik verisini giriniz: "))
            try:
                baglanti.execute("INSERT INTO Fransa(Enflasyon,Issizlik) VALUES ('{}','{}')".format(enflasyon,issizlik))
                baglanti.commit()
            except:
                print("Veritabanı bağlantısında bir hata oluştu.")
            finally:
                print("Ekleme Başarılı")
            ay = ay + 1
            i = i + 1
    def ukraine():
        i = 0 
        ay = 1
        while i < 13:
            enflasyon = int(input(f"{ay}. Ayın enflasyon verisini giriniz: "))
            issizlik = int(input(f"{ay}. Ayın işsizlik verisini giriniz: "))
            try:
                baglanti.execute("INSERT INTO Ukrayna(Enflasyon,Issizlik) VALUES ('{}','{}')".format(enflasyon,issizlik))
                baglanti.commit()
            except:
                print("Veritabanı bağlantısında bir hata oluştu.")
            finally:
                print("Ekleme Başarılı")
            ay = ay + 1
            i = i + 1
    def spain():
        i = 0 
        ay = 1
        while i < 13:
            enflasyon = int(input(f"{ay}. Ayın enflasyon verisini giriniz: "))
            issizlik = int(input(f"{ay}. Ayın işsizlik verisini giriniz: "))
            try:
                baglanti.execute("INSERT INTO Spain(Enflasyon,Issizlik) VALUES ('{}','{}')".format(enflasyon,issizlik))
                baglanti.commit()
            except:
                print("Veritabanı bağlantısında bir hata oluştu.")
            finally:
                print("Ekleme Başarılı")
            ay = ay + 1
            i = i + 1
    def veriekle():
        secim = input("Hangi ülkeye veri eklemek istiyorsunuz: \n 1-) ABD \n 2-) Birleşik Krallık \n 3-) Fransa \n 4-) Ukrayna \n 5-) İspanya \n ")
        if secim == "1":
            ulkeler.abd()
        elif secim == "2":
            ulkeler.england()
        elif secim == "3":
            ulkeler.france()
        elif secim == "4":
            ulkeler.ukraine()
        elif secim == "5":
            ulkeler.spain()
        else:
            print("Hatalı bir tuşlama yaptınız.")
    
    

class karsilastirma: 
    def gorus():
        secim = input("Hangi ülke ilgili görüş belirtmek istiyorsunuz: 1-) ABD \n 2-) Birleşik Krallık \n 3-) Fransa \n 4-) Ukrayna \n 5-) İspanya \n")
        if secim == "1":
            gorus = input("ABD hakkındaki görüşleriniz: ")
            dosya = open("abd.txt","w")
            dosya.write(gorus)
            print("Görüşleriniz dosyaya eklendi.")
            dosya.close()
        elif secim == "2":
            gorus = input("Birleşik Krallık hakkındaki görüşleriniz: ")
            dosya = open("england.txt","w")
            dosya.write(gorus)
            print("Görüşleriniz dosyaya eklendi.")
            dosya.close()
        elif secim == "3":
            gorus = input("Fransa hakkındaki görüşleriniz: ")
            dosya = open("france.txt","w")
            dosya.write(gorus)
            print("Görüşleriniz dosyaya eklendi.")
            dosya.close()
        elif secim == "4":
            gorus = input("İspanya hakkındaki görüşleriniz: ")
            dosya = open("spain.txt","w")
            dosya.write(gorus)
            print("Görüşleriniz dosyaya eklendi.")
            dosya.close()
        elif secim == "5":
            gorus = input("Ukrayna hakkındaki görüşleriniz: ")
            dosya = open("ukraine.txt","w")
            dosya.write(gorus)
            print("Görüşleriniz dosyaya eklendi.")
            dosya.close()
    def enflasyon():
        plt.plot(abd, color = "blue", label = "Abd Enflasyon")
        plt.plot(england, color = "red", label = "Birleşik Krallık Enflasyon")
        plt.plot(france, color = "pink", label = "Fransa Enflasyon")
        plt.plot(spain, color = "orange", label = "İspanya Enflasyon")
        plt.plot(ukraine, color = "purple", label = "Ukrayna Enflasyon")
        plt.title("5 Ülkenin 2022 - Enflasyon Grafiği ")
        plt.xlabel("Aylar")
        plt.ylabel("Oranlar")
        plt.legend()
        plt.show()
        karsilastirma.gorus()
    def issizlik():
        plt.plot(abd1, color = "blue" , label = "Abd İşsizlik")
        plt.plot(england1, color = "red" , label = "Birleşik Krallık İşsizlik")
        plt.plot(france1, color = "pink" , label = "Fransa İşsizlik")
        plt.plot(spain1, color = "orange" , label = "İspanya İşsizlik")
        plt.plot(ukraine1, color = "purple" , label = "Ukrayna İşsizlik")
        plt.title("5 Ülkenin 2 yıllık İşsizlik Grafiği ")
        plt.xlabel("Aylar")
        plt.ylabel("Oranlar")
        plt.legend()
        plt.show()
        karsilastirma.gorus()
    def ikili():
        plt.plot(abd, color = "blue", label = "Abd Enflasyon")
        plt.plot(abd1, color = "blue" , label = "Abd İşsizlik")
        plt.plot(england, color = "red", label = "Birleşik Krallık Enflasyon")
        plt.plot(england1, color = "red" , label = "Birleşik Krallık İşsizlik")
        plt.plot(france, color = "pink", label = "Fransa Enflasyon")
        plt.plot(france1, color = "pink" , label = "Fransa İşsizlik")
        plt.plot(spain, color = "orange", label = "İspanya Enflasyon")
        plt.plot(spain1, color = "orange" , label = "İspanya İşsizlik")
        plt.plot(ukraine, color = "purple", label = "Ukrayna Enflasyon")
        plt.plot(ukraine1, color = "purple" , label = "Ukrayna İşsizlik")
        plt.title("5 Ülkenin 2022 Enflasyon ve 2 Yıllık İşsizlik Grafiği ")
        plt.xlabel("Aylar")
        plt.ylabel("Oranlar")
        plt.legend()
        plt.show()
        karsilastirma.gorus()
    def secim():
        secim = input("Ülkeleri hangi ölçekte kıyaslamak istiyorsunuz:  \n 1-) Enflasyon \n 2-) İşsizlik \n 3-) İkiside \n ")
        if secim == "1":
            karsilastirma.enflasyon()
        elif secim == "2":
            karsilastirma.issizlik()
        elif secim == "3":
            karsilastirma.ikili()
    
def Anamenu():
    print("-" * 50, "Ülkeler Sistemine Hoşgeldiniz", "-" * 50)
    sec = input("Ne yapmak istiyorsunuz: \n 1-) Ülke karşılaştırmak \n 2-) Ülkelere veri eklemek \n")
    if sec == "1":
        karsilastirma.secim()
    elif sec == "2":
        ulkeler.veriekle()
    else:
        print("Hatalı bir tuşlama yaptınız.")
    
Anamenu()
        
```

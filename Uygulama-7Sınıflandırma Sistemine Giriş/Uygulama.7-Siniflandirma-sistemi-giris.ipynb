```python
import numpy as np
import pandas as pd
from numpy import random
    
dosya = pd.read_csv("kart.csv")

kart = dosya["Kart"].values

def kartcek():
    i= 0 
    while i<5:
        x = random.randint(100, size=1)
        for a in kart:
            for b in x:
                if b == a:
                    print(f"Tebrikler bir sayı tuttu \n Çekilen Sayı: {a} = {b}")
                else:
                    print(f"Üzgünüz bir sayı tutmadı \n Çekilen Sayı: {a} ")
        i= i + 1
   
kartcek()
```

```python
import matplotlib.pyplot as plt
import numpy as np

g = 10

def grafik():
    h = np.array(['2','43','21','10','75'])
    Ep = np.array(['152','354','250','145','176'])
    plt.title("Potansiyel Enerji - Yükseklik Grafiği")
    plt.xlabel("Yükseklik")
    plt.ylabel("Potansiyel Enerji")
    plt.plot(h,Ep)
    plt.show()
    
def pot():
    m = int(input("m değerini giriniz: "))
    h = int(input("h değerini giriniz: "))
    potansiyel = m * g * h
    secims = input("Çıkarım yapmak ister misiniz (E/H): \n")
    if secims == "E" or secims == "e":
        grafik()
        print(f"Potansiyel enerji: {potansiyel}")
    elif secims == "H" or secims == "h":
        print(f"Potansiyel enerji: {potansiyel}")
    else:
        print("Yanlış bir tuşa tıkladınız.")
        
        
def grafiks():
    kısa = np.array(['20','43','21','10','75'])
    alan = np.array(['100','132','125','214','156'])
    plt.title("Dikdörtgen - kısa ve uzun kenar grafiği")
    plt.xlabel("Kısa Kenar")
    plt.ylabel("Alan")
    plt.plot(kısa,alan, "r-")
    plt.show()

def dıkdortgen():
    kısa = int(input("Kısa kenar değerini giriniz: "))
    uzun = int(input("Uzun kenar değerini giriniz: "))
    alan = kısa * uzun
    secims = input("Çıkarım yapmak ister misiniz (E/H): \n")
    if secims == "E" or secims == "e":
        grafiks()
        print(f"Dikdörtgenin Alanı: {alan}")
    elif secims == "H" or secims == "h":
        print(f"Dikdörtgenin Alanı: {alan}")
    else:
        print("Yanlış bir tuşa tıkladınız.")
    
print("-" * 50, "Yeni Programamıza Hoşgeldiniz: ", "-" * 50 )
secim = input("Ne yapmak istiyorsunuz: \n 1-)Potansiyel Enerji Hesaplama: \n 2-) Dikdörtgenin Alanını Hesapla: \n ")


if secim == "1":
    pot()
elif secim == "2":
    dıkdortgen()
else:
    pass
```
